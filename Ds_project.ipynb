Importing modules/libraries: Pandas, Numpy, Seaborn, Matplotlib

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Importing the dataset which is in csv format.
The dataset is downloaded from kaggle.

```python
data = pd.read_csv('/content/world-happiness-report-2021.csv')
```

## Part 1. Data Preprocessing
This will involve viewing our dataset in tabular form and then ensuring that the **formatting of the data is consistent** known as **data cleaning/scrubbing** (deleting inaccurate and duplicate values in dataset, handling any gaps in the dataset)

To view our dataset using the head and tail functions which show us the 10 rows from the start and end respectively by default.

```python
data.head()
```

```python
data.tail()
```

Keeping only the relevant columns for our data analysis.

```python
data_columns = ['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
data = data[data_columns].copy() # coping only the columns we need 
data.head()
```

Editing the column names

```python
data.columns = ['country_name','regional_indicator','ladder_score','logged_GDP_per_capita','social_support','healthy_life_expectancy','freedom_to_make_life_choices','generosity','perceptions_of_corruption']
data.head()
```

Checking the number of null columns in our dataset.

If any value is find null we can replace it with the average of that column or drop that particular row.

```python
data.isnull().sum()
```

There are no null columns in our dataset hence there is no need for cleaning the dataset.

## Part 2. Data Vizualization
Here we will graphically represent our data using visual elements like charts, graphs, and maps. 
For this purpose we use the inbuilt functions from Matplotlib and Seaborn.

Setting the proper parameters for data vizualization.

```python
sns.set_style('darkgrid') 
plt.rcParams['font.size'] = 15 
plt.rcParams['figure.figsize'] = (10,7)  # default size of all figure
plt.rcParams['figure.facecolor'] = '#EFFAFB'  # default background color
```

Plot between happiness and gdp


```python
plt.rcParams['figure.figsize'] = (12,7)
plt.title('Plot between Happiness score and GDP')
sns.scatterplot(x = data.ladder_score, y = data.logged_GDP_per_capita,hue = data.regional_indicator, s = 200);

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Ladder score')
plt.ylabel('Logged GDP per capita') 
```

In the above graph we observe that countries from western Europe have highest happiness score and highest GDP per capita.
Afganistan has the lowest happiness score.

Further, to find which country has the highest contribution in world's GDP, we will use pie plot to vizualize the GDP by region.


```python
gdp_region = data.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
print(gdp_region)
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP by Region')
plt.ylabel('')
```

From the above pie plot we observe that Sub-Saharan Africa and Western Europe are contributing highest GDP where as North America and ANZ are contributing the least to the world GDP.

Now we will count number of countries in each region using the groupby function:

```python
total_country = data.groupby('regional_indicator')[['country_name']].count()
print(total_country)
```

Plotting the correlation map to understand the relation between the features:

```python
cor = data.corr(method = 'pearson')
f, ax = plt.subplots(figsize = (10,5))
sns.heatmap(cor, mask = np.zeros_like(cor, dtype=np.bool), cmap = 'Blues')
```

In the above heatmap the dark blue region indicates the highest correlation and the white grey color indicates the low correlation. 

Here we observe that there is negative correlation between happiness_score and the percepton of corruption since if the citizens of the country feel that there is lot of corruption in the country their happiness score would be low.

**Analysing the corruption** in different regions using the groupby function:

```python
corruption = data.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruption
```

From the above table we observe that North America and ANZ and Western Europe have the lowest corruption

Now we will visualize this data.

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('Perception of corruption in various regions')
plt.xlabel('Regions', fontsize = 15)
plt.ylabel('Corruption Index', fontsize = 15)
plt.xticks(rotation = 30)
plt.bar(corruption.index, corruption.perceptions_of_corruption)
```

Now we will fetch the top 10 high life expectancy countries and bottom 10 and visualize them.

```python
top_10 = data.head(10)
bottom_10 = data.tail(10)  
```

```python
fig, axes = plt.subplots(1,2, figsize = (16,6))
plt.tight_layout(pad = 2)
xlabels = top_10.country_name
axes[0].set_title('Top 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels , rotation=45, ha = 'right')
sns.barplot(x = top_10.country_name, y = top_10.healthy_life_expectancy, ax = axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life Expectancy')


xlabels = bottom_10.country_name
axes[1].set_title('Bottom 10 happiest countries Life Expectancy')
axes[1].set_xticklabels(xlabels , rotation=45, ha = 'right')  # ha is horizontal allignment
sns.barplot(x = bottom_10.country_name, y = bottom_10.healthy_life_expectancy, ax = axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life Expectancy')
```

Now we will vizualize the relationship between freedom to make life choices and the happiness score using scatterplot from seaborn library.

```python
plt.rcParams['figure.figsize'] = (15,7)
sns.scatterplot(x = data.freedom_to_make_life_choices, y = data.ladder_score, hue = data.regional_indicator, s = 200)
plt.legend(loc = 'upper left', fontsize = '12')
plt.xlabel('Freedom to Make Choices')
plt.ylabel('Happiness Score')
```

We observe that in western countries, the freedom to make life choices is higher so the happiness score is also high. And in the middle east africa the freedom to make life choices is lower so the happiness score in that region is also low.

In Southeast Asian countries even though the happiness score is comparitively lower but the freedom to make life choices is more than 0.8.

In South Asia i.e Afganisthan the happiness score as well as the freedom to make life choices are lower.

Top 10 corruption countries

```python
country = data.sort_values(by = 'perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Countries with Most Perception of Corruption ')
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Corruption Index', fontsize = 13)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.country_name, country.perceptions_of_corruption)
```

Now we will visualize the countries with **highest perception of corruption** using a bar graph from matplotlib.

```python
country = data.sort_values(by = 'perceptions_of_corruption').tail(10)
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Countries with Most Perceprion of Corruption ')
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Corruption Index', fontsize = 13)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.country_name, country.perceptions_of_corruption)
```

Now we can visualize **how the corruption varies in terms of happiness score**.

```python
plt.rcParams['figure.figsize'] = (15,7)
sns.scatterplot(x = data.ladder_score, y = data.perceptions_of_corruption, hue = data.regional_indicator, s = 200)
plt.legend(loc = 'lower left', fontsize = '14')
plt.xlabel('Happiness Score')
plt.ylabel('Corruption')
```

Hence, we can infer that the countries with a low corruption have a high happiness score, for example - western europe.
